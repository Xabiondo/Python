```python
titanic_pasajeros_mean = titanic_pasajeros.Fare.mean()
```

```python
print(titanic_pasajeros_mean)
```

```python
def function_lambda(p):
    p - titanic_pasajeros_mean
    return p
```

```python
titanic_pasajeros.Fare.map(function_lambda )
```

```python
titanic_pasajeros.Fare.map(lambda  p : p - titanic_pasajeros_mean)
```

```python
titanic_pasajeros.Fare.map(titanic_pasajeros['Fare'] - titanic_pasajeros_mean)
```

```python
def remean_points(row):
    row.Fare = row.Fare - titanic_pasajeros_mean
    return row
```

```python
titanic_pasajeros.apply(remean_points , axis='columns')
#Esto quiere decir , que recorre las columnas horizontalemtene, es decir
#Que esta recorriendo una fila jajaj,  y al revés
#Esto me cuesta entenderlo, pero básicamente, coje toda la fila, como si fuera un objeto, y le pasa la función que tu le pongas como primer parámetro .
#Esto modifica una copia , es paso por valor , no por referencia
```
